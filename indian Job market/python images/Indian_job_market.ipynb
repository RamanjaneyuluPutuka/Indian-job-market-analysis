```python
df.columns=df.columns.str.replace(" ","_")
df.columns=df.columns.str.replace("/","_")
```

```python
df['experience_required']=df['experience_required'].str.replace("years","",regex=False)
df['experience_required']=df['experience_required'].astype(str)
```

```python
df.info()
```

```python
#Creating a column Grouping
def Agg(education_requirement):
  if "0-2" in education_requirement:
    return "L1"
  elif "2-5" in education_requirement:
    return "L2"
  elif "5-10" in education_requirement:
    return "L3"
  elif "10+" in education_requirement:
    return "L4"
  else:
    return "Unknown"

df['experience_level']=df['experience_required'].apply(Agg)

```
